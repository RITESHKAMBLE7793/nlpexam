#  09 :Implementation of Agglomerative Hierarchical Clustering for Wine Quality Analysis

# Step 1: Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
```

# Step 2: Load Dataset

```python
data = pd.read_csv("winequality-red.csv", sep=';')
```

```python
data
```

# Step 3: View Dataset

```python
print(data.head())
print(data.columns)
```

# Step 4: Separate Features

```python
X = data.drop("quality", axis=1)
```

# Step 5: Normalize Data

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# Step 6: Apply Agglomerative Clustering

```python
model = AgglomerativeClustering(n_clusters=3)

labels = model.fit_predict(X_scaled)
```

# Step 7: Add Cluster to Dataset

```python
data['Cluster'] = labels
print(data.head())
```

# Step 8: Check Cluster Values

```python
print("Clusters:", data['Cluster'].unique())
```

# Step 9: Count Wines in Each Cluster

```python
print(data['Cluster'].value_counts())
```

# Step 10: Visualize Clusters

```python
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels)
plt.title("Wine Clusters")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
Short Notes – Agglomerative Hierarchical Clustering
Agglomerative clustering is an unsupervised learning technique.
It groups data points based on similarity.
Starts with each data point as a separate cluster and merges them step by step.

Dataset
Wine Quality Dataset (UCI Repository)
File used: winequality-red.csv

Steps
Load dataset (sep=';')
Remove quality column
Apply data scaling
Apply AgglomerativeClustering
Assign cluster labels

Output
New column → Cluster (0,1,2)
Wines are grouped into similar clusters
Scatter plot shows different groups


Why we are doing this practical
    The main purpose is to understand how data can be grouped automatically without labels.
In Simple Words
To make the computer:
find similar wines  and group them into clusters
```
