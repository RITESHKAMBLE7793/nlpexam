# 06 :Implementation of Word Sense Disambiguation (WSD) using WordNet

# Step 1: Install & Download

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
```

# Step 2: Import Libraries

```python
from nltk.corpus import wordnet as wn
```

# Step 3: Input Word

```python
word = "bank"
```

# Step 4: Get All Meanings

```python
synsets = wn.synsets(word)

for s in synsets:
    print(s.name(), ":", s.definition())
```

# Step 5: Simple WSD (Context Matching)

```python
context = "He went to the bank to deposit money"

best_sense = None
max_overlap = 0

for sense in synsets:
    definition = sense.definition()
    overlap = len(set(context.split()) & set(definition.split()))

    if overlap > max_overlap:
        max_overlap = overlap
        best_sense = sense

print("\nBest Sense:", best_sense)
print("Meaning:", best_sense.definition())
```

```python
What is the Aim (in very simple words)

👉 The aim is to find the correct meaning of a word when it has multiple meanings.

🔹 Example to understand

Word: bank

It has 2 meanings:

Money bank (financial place)

River bank (side of river)

✅ What your program does

It checks the sentence (context) and decides:

👉 Which meaning is correct?
```
