# Step 2: Import Library

```python
import spacy
```

# Step 3: Load Model

```python
nlp = spacy.load("en_core_web_sm")
```

# Step 4: Input Text

```python
text = "Riya has a dog. She loves it."
doc = nlp(text)
```

# Step 5: Identify Pronouns and Nouns

```python
for token in doc:
    if token.pos_ == "PRON":
        print("Pronoun:", token.text)
    if token.pos_ == "NOUN" or token.pos_ == "PROPN":
        print("Noun:", token.text)
```

# Step 6: Simple Co-reference Logic

```python
entities = []
for token in doc:
    if token.pos_ in ["NOUN", "PROPN"]:
        entities.append(token.text)

for token in doc:
    if token.pos_ == "PRON":
        print(token.text, "→ refers to →", entities[-1])
```

```python
Purpose of the Practical

The purpose of this practical is to understand how a computer identifies relationships between words in a sentence, specifically how pronouns (like he, she, it) refer to actual entities (like names or objects).

In Simple Words

To make the system understand:
👉 “Who is who” in a sentence

Example:
Riya has a dog. She loves it.

She → Riya

it → dog

Why this is important (significance)

Improves text understanding in NLP

Used in chatbots and virtual assistants

Helps in document summarization

Important for question answering systems
```
