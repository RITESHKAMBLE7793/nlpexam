# Step 2: Import Libraries

```python
import nltk
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

# Step 3: Download NLTK Data

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

# Step 4: Load Dataset (Sample Text)

```python
text = "This movie was AMAZING! I loved the acting, but the story was a bit slow."
```

# Step 5: Convert to Lowercase (Normalization)

```python
text = text.lower()
print("Lowercase Text:", text)
```

# Step 6: Remove Special Characters

```python
text = re.sub(r'[^a-zA-Z\s]', '', text)
print("Clean Text:", text)
```

# Step 7: Tokenization

```python
tokens = word_tokenize(text)
print("Tokens:", tokens)
```

# Step 8: Remove Stopwords

```python
stop_words = set(stopwords.words('english'))

filtered_tokens = [word for word in tokens if word not in stop_words]
print("After Stopword Removal:", filtered_tokens)
```

# Step 9: Stemming

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()
stemmed_words = [ps.stem(word) for word in filtered_tokens]

print("Stemmed Words:", stemmed_words)
```
