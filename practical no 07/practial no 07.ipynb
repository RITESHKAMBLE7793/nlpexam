```python
pip install gensim
```

# Step 2: Import Libraries

```python
from gensim.models import Word2Vec
from gensim.downloader import load
```

# Step 3: Load Dataset (Text8)

```python
dataset = load("text8")
```

# Step 4: Train Word2Vec Model

```python
model = Word2Vec(dataset, vector_size=100, window=5, min_count=5, workers=4)
```

# Step 5: Check Word Vector

```python
print(model.wv["king"])
```

 # Step 6: Find Similar Words

```python
print(model.wv.most_similar("king"))
```

# Step 7: Analogy

```python
print(model.wv.most_similar(positive=["king", "woman"], negative=["man"]))
```

```python
Short Notes – Word Embedding (Word2Vec)

Word Embedding converts words into numerical vectors.

It learns meaning based on context of words in a dataset.

Words with similar meaning have similar vectors.

Working

Model is trained on a text dataset (Text8)

It learns relationships between words

Uses Word2Vec algorithm

Output Understanding

Word Vector → numerical representation of word

Similar Words → words with similar meaning

Analogy → relationship between words

Example: king - man + woman = queen

Significance

Helps in understanding text

Used in NLP tasks like translation, chatbots

Conclusion

Word2Vec represents words as vectors and captures their meaning and relationships.
```
