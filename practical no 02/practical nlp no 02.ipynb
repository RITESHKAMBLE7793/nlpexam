# 02 : Spell Checking and Correction using Minimum Edit Distance

# Step 1: Define Minimum Edit Distance Function

```python
def edit_distance(str1, str2):
    m = len(str1)
    n = len(str2)

    # Create matrix
    dp = [[0 for _ in range(n+1)] for _ in range(m+1)]

    # Initialize
    for i in range(m+1):
        dp[i][0] = i
    for j in range(n+1):
        dp[0][j] = j

    # Compute distance
    for i in range(1, m+1):
        for j in range(1, n+1):
            if str1[i-1] == str2[j-1]:
                cost = 0
            else:
                cost = 1

            dp[i][j] = min(
                dp[i-1][j] + 1,      # Deletion
                dp[i][j-1] + 1,      # Insertion
                dp[i-1][j-1] + cost  # Substitution
            )

    return dp[m][n]
```

# Step 2: Spell Correction Logic

```python
def correct_word(input_word, dictionary):
    min_dist = float('inf')
    correct = ""

    for word in dictionary:
        dist = edit_distance(input_word, word)
        
        if dist < min_dist:
            min_dist = dist
            correct = word

    return correct, min_dist
```

# Step 3: Input Word and Dataset

```python
dictionary = ["book", "back", "cook", "look", "took"]
input_word = "toak"
```

# Step 4: Perform Spell Correction

```python
corrected_word, distance = correct_word(input_word, dictionary)

print("Input Word:", input_word)
print("Corrected Word:", corrected_word)
print("Edit Distance:", distance)
```
